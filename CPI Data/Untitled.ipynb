```python
"""
CPI DRIVERS – CANADA VS US
--------------------------

Goal:
    Quantify which components of the Consumer Price Index (CPI) are actually
    driving inflation in the most recent month for:
        - Canada (by major CPI components)
        - United States (headline CPI-U Table 1 categories)

    For each country we:
        1. Use CPI basket weights as "importance" (share in the index).
        2. Combine weights with monthly price changes to get:
              contribution_pp = weight * monthly % change
           -> tells us how many percentage points each component contributes
              to the overall CPI move.
        3. Compute each component’s share of the total CPI move.
        4. Visualize contributions with bar charts, pie charts, and a Canada–US
           comparison for Food & Shelter.

    For Canada we also:
        - Compute YEAR-ON-YEAR (YoY) contributions (Sep 2024 → Sep 2025).
        - Compare MoM vs YoY contributions by component.
        - Classify components as:
            * "Persistent upward pressure"
            * "Upward pressure but easing"
            * "Short-term upside shock"
            * "Disinflation / minor"
        - Plot a MoM vs YoY scatter to visually separate temporary vs persistent moves.

Data Sources (local CSVs you downloaded):
    - CAN_WEIGHTS_PATH:
        2024 CPI basket weights by major component for Canada
        (Statistics Canada "Basket weights of the CPI", Canada, 2024)

    - CAN_INDEX_PATH:
        Monthly CPI index by component for Canada, May–Sep 2025
        (Statistics Canada CPI by product group – exported subset to CSV)

    - CAN_INDEX_YOY_PATH:
        CPI index by component with both September 2024 and September 2025
        (same StatCan table but with a wider time span; YoY data)

    - US_TABLE1_PATH:
        US CPI for All Urban Consumers (CPI-U), Table 1, Sep 2025 news release
        (Bureau of Labor Statistics, "news-release-table1-202509.csv")

Make sure the BASE_DIR below points to the folder where you saved these CSV files.
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

# -------------------------------------------------------------------
# 0. Paths / plotting style / month configuration
# -------------------------------------------------------------------
plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams["axes.grid"] = True

# Root folder where your CPI CSVs live
BASE_DIR = "/Users/rjoshi_24/Desktop/CPI Data/Data"  # <-- change if needed

# Exact file names (already saved in that folder)
CAN_WEIGHTS_PATH    = f"{BASE_DIR}/2024 CPI.csv"
CAN_INDEX_PATH      = f"{BASE_DIR}/1810000401-noSymbol.csv"
CAN_INDEX_YOY_PATH  = f"{BASE_DIR}/1810000401-noSymbol (2).csv"  # YoY file
US_TABLE1_PATH      = f"{BASE_DIR}/news-release-table1-202509.csv"

# Which month-over-month move we focus on for Canada
FROM_MONTH_CAN = "August 2025"
TO_MONTH_CAN   = "September 2025"

# Which year-on-year move (columns in the YoY file) we focus on for Canada
YOY_FROM = "September 2024"
YOY_TO   = "September 2025"

# ===================================================================
# CANADA SECTION – WEIGHTS & MONTHLY INDEX
# ===================================================================

# -------------------------------------------------------------------
# 1. Load 2024 CPI basket weights for Canada (major components only)
# -------------------------------------------------------------------
can_w = pd.read_csv(CAN_WEIGHTS_PATH)

# Rename columns to something simple
can_w.columns = ["component", "weight_2024"]

# Clean spacing in component names so they match other files
can_w["component"] = (
    can_w["component"]
    .astype(str)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

# Focus on the 8 high-level CPI components (ignore All-items, sub-subcategories)
major_components = [
    "Food",
    "Shelter",
    "Household operations, furnishings and equipment",
    "Clothing and footwear",
    "Transportation",
    "Health and personal care",
    "Recreation, education and reading",
    "Alcoholic beverages, tobacco products and recreational cannabis",
]
can_w = can_w[can_w["component"].isin(major_components)].copy()

# Convert weights from percent (e.g. 16.9) to fraction of the basket (0.169)
can_w["weight_frac"] = can_w["weight_2024"] / 100.0

print("Canada weights (2024 basket, major components):")
display(can_w)

# -------------------------------------------------------------------
# 2. Load monthly CPI index by component (Canada, May–Sep 2025)
# -------------------------------------------------------------------
raw_can_idx = pd.read_csv(CAN_INDEX_PATH, skiprows=7)

# First row after skip is actually the header row
new_header = raw_can_idx.iloc[0]
can_idx = raw_can_idx.iloc[1:].copy()
can_idx.columns = new_header

# Keep only rows where the product group name is present
can_idx = can_idx[can_idx["Products and product groups 3 4"].notna()].copy()

# Rename the product-group column to match our weights DataFrame
can_idx = can_idx.rename(columns={"Products and product groups 3 4": "component"})

# Clean up component names: drop trailing footnote numbers and fix spacing
can_idx["component"] = (
    can_idx["component"]
    .astype(str)
    .str.replace(r"\s+\d+$", "", regex=True)  # remove footnote numbers
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

# Monthly columns available in this file
month_cols = ["May 2025", "June 2025", "July 2025", "August 2025", "September 2025"]
can_idx = can_idx[["component"] + month_cols].copy()

# Convert to numeric
for col in month_cols:
    can_idx[col] = pd.to_numeric(can_idx[col], errors="coerce")

print("Canada CPI index by component (first 10 rows):")
display(can_idx.head(10))

# ===================================================================
# CANADA – MoM CONTRIBUTIONS (FROM_MONTH_CAN → TO_MONTH_CAN)
# ===================================================================

# -------------------------------------------------------------------
# 3. Compute Canadian contributions for the selected month (FROM → TO)
# -------------------------------------------------------------------
can_combined = can_idx.merge(
    can_w[["component", "weight_frac"]],
    on="component",
    how="inner"
)

latest_col = TO_MONTH_CAN
prev_col   = FROM_MONTH_CAN

# Monthly % change in the CPI index for each component
can_combined["monthly_pct_change"] = (
    (can_combined[latest_col] / can_combined[prev_col] - 1.0) * 100.0
)

# Contribution in percentage points:
#   contribution_pp = weight (fraction) * monthly % change
can_combined["contribution_pp"] = (
    can_combined["weight_frac"] * can_combined["monthly_pct_change"]
)

# How much of the total CPI move does each component explain?
total_contrib_can = can_combined["contribution_pp"].sum()
if abs(total_contrib_can) < 1e-9:
    print("Total Canadian CPI move is ~0; shares would be unstable.")
    can_combined["share_of_total_%"] = np.nan
else:
    can_combined["share_of_total_%"] = (
        can_combined["contribution_pp"] / total_contrib_can * 100.0
    )

can_combined_sorted = can_combined.sort_values("contribution_pp", ascending=False)

print(f"Canada – contributions to CPI change ({prev_col} → {latest_col}):")
display(can_combined_sorted)

# -------------------------------------------------------------------
# 3a. Canada – headline summary + sanity check
# -------------------------------------------------------------------
# Use the 'All-items' row from can_idx to get the official headline move.
all_items_can = can_idx[can_idx["component"] == "All-items"].iloc[0]
headline_change_can = (
    (all_items_can[latest_col] / all_items_can[prev_col] - 1.0) * 100.0
)

top3_can = can_combined_sorted.head(3)
top3_share_can = top3_can["share_of_total_%"].sum()

print(
    f"Canada headline CPI moved {headline_change_can:.2f}% m/m "
    f"from {prev_col} to {latest_col}."
)
print("Top 3 Canadian drivers:")
for _, row in top3_can.iterrows():
    print(
        f"  - {row['component']}: "
        f"{row['contribution_pp']:.3f} p.p. "
        f"({row['share_of_total_%']:.1f}% of the move)"
    )
print(f"Top 3 explain {top3_share_can:.1f}% of the total move.\n")

print(
    "Sanity check – sum of component contributions "
    f"(subset of basket): {total_contrib_can:.3f} p.p."
)
print(
    f"Headline CPI move from All-items index: {headline_change_can:.3f} p.p.\n"
)

# -------------------------------------------------------------------
# 4. Canada – bar chart: horizontal, cleaner layout
# -------------------------------------------------------------------
plot_df = can_combined_sorted.sort_values("contribution_pp")  # smallest at bottom
y = np.arange(len(plot_df))

fig, ax = plt.subplots(figsize=(12, 7))
bars = ax.barh(y, plot_df["contribution_pp"])

ax.set_yticks(y)
ax.set_yticklabels(plot_df["component"], fontsize=9)
ax.set_xlabel("Contribution to monthly CPI change (percentage points)", fontsize=11)
ax.set_title(
    f"Canada: CPI component contributions, {prev_col} → {latest_col}",
    fontsize=16,
    pad=15,
)

ax.tick_params(axis="x", labelsize=10)
ax.axvline(0, color="black", linewidth=0.8)

# Label bars with share-of-move on the right side
for bar, share in zip(bars, plot_df["share_of_total_%"]):
    if pd.isna(share):
        continue
    width = bar.get_width()
    x_pos = width + (0.01 if width >= 0 else -0.01)
    va = "center"
    ha = "left" if width >= 0 else "right"
    ax.text(
        x_pos,
        bar.get_y() + bar.get_height() / 2,
        f"{share:.1f}%",
        va=va,
        ha=ha,
        fontsize=9,
    )

plt.tight_layout()
plt.show()

# -------------------------------------------------------------------
# 5. Canada – pie chart: which components drove the CPI *increase*?
# -------------------------------------------------------------------
positive_can = can_combined[can_combined["contribution_pp"] > 0].copy()
positive_can = positive_can.sort_values("share_of_total_%", ascending=False)

fig, ax = plt.subplots(figsize=(9, 9))
wedges, texts, autotexts = ax.pie(
    positive_can["share_of_total_%"],
    labels=positive_can["component"],
    autopct="%1.1f%%",
    startangle=90,
    counterclock=False,
    textprops={"fontsize": 10},
)

for txt in texts:
    txt.set_fontsize(9)
for at in autotexts:
    at.set_fontsize(10)

ax.set_title(
    "Canada: share of CPI increase by component\n"
    f"(positive contributors, {prev_col} → {latest_col})",
    fontsize=14,
)
plt.show()

# -------------------------------------------------------------------
# 6. Canada – contributions over time (May→Jun, Jun→Jul, Jul→Aug, Aug→Sep)
# -------------------------------------------------------------------
month_pairs = list(zip(month_cols[:-1], month_cols[1:]))

records = []
for pm, cm in month_pairs:
    tmp = can_idx.merge(
        can_w[["component", "weight_frac"]],
        on="component",
        how="inner"
    ).copy()
    tmp["monthly_pct_change"] = (tmp[cm] / tmp[pm] - 1.0) * 100.0
    tmp["contribution_pp"] = tmp["weight_frac"] * tmp["monthly_pct_change"]

    for _, row in tmp.iterrows():
        records.append({
            "period": f"{pm} → {cm}",
            "component": row["component"],
            "contribution_pp": row["contribution_pp"],
        })

can_contrib_long = pd.DataFrame(records)
can_contrib_pivot = can_contrib_long.pivot(
    index="period",
    columns="component",
    values="contribution_pp"
)

print("Canada – contributions over time (pivot table):")
display(can_contrib_pivot)

focus_components = ["Shelter", "Food", "Transportation"]

ax = can_contrib_pivot[focus_components].plot(
    kind="bar", figsize=(12, 6)
)
plt.ylabel("Contribution to monthly CPI change (percentage points)", fontsize=11)
plt.title("Canada: contributions over time (selected components)", fontsize=14)
plt.xticks(rotation=45, ha="right", fontsize=9)
plt.yticks(fontsize=10)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()

# ===================================================================
# CANADA – YEAR-ON-YEAR (YoY) CONTRIBUTIONS (Sep 2024 → Sep 2025)
# ===================================================================

raw_can_idx_yoy = pd.read_csv(CAN_INDEX_YOY_PATH, skiprows=7)

new_header_yoy = raw_can_idx_yoy.iloc[0]
can_idx_yoy = raw_can_idx_yoy.iloc[1:].copy()
can_idx_yoy.columns = new_header_yoy

can_idx_yoy = can_idx_yoy[can_idx_yoy["Products and product groups 3 4"].notna()].copy()
can_idx_yoy = can_idx_yoy.rename(columns={"Products and product groups 3 4": "component"})

can_idx_yoy["component"] = (
    can_idx_yoy["component"]
    .astype(str)
    .str.replace(r"\s+\d+$", "", regex=True)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

# Keep only Sept 2024 & Sept 2025
can_idx_yoy = can_idx_yoy[["component", YOY_FROM, YOY_TO]].copy()

can_idx_yoy[YOY_FROM] = pd.to_numeric(can_idx_yoy[YOY_FROM], errors="coerce")
can_idx_yoy[YOY_TO]   = pd.to_numeric(can_idx_yoy[YOY_TO], errors="coerce")

# Merge with weights
can_yoy = can_idx_yoy.merge(
    can_w[["component", "weight_frac"]],
    on="component",
    how="inner"
)

# YoY % change
can_yoy["yoy_pct_change"] = (
    (can_yoy[YOY_TO] / can_yoy[YOY_FROM] - 1.0) * 100.0
)

# YoY contribution in percentage points
can_yoy["yoy_contribution_pp"] = (
    can_yoy["weight_frac"] * can_yoy["yoy_pct_change"]
)

# Share of total YoY CPI move (within our subset)
total_yoy_contrib = can_yoy["yoy_contribution_pp"].sum()
if abs(total_yoy_contrib) < 1e-9:
    print("Total Canadian YoY CPI move is ~0; YoY shares unstable.")
    can_yoy["yoy_share_of_total_%"] = np.nan
else:
    can_yoy["yoy_share_of_total_%"] = (
        can_yoy["yoy_contribution_pp"] / total_yoy_contrib * 100.0
    )

can_yoy_sorted = can_yoy.sort_values("yoy_contribution_pp", ascending=False)

print(f"Canada – YoY contributions to CPI change ({YOY_FROM} → {YOY_TO}):")
display(can_yoy_sorted)

# ===================================================================
# CANADA – MoM vs YoY: classify persistent vs temporary
# ===================================================================

combined_mom_yoy = can_combined.merge(
    can_yoy[["component", "yoy_contribution_pp", "yoy_share_of_total_%"]],
    on="component",
    how="inner",
    suffixes=("_mom", "_yoy")
)

def classify_row(r, mom_thresh=0.02, yoy_thresh=0.10):
    """
    Simple classification of components based on MoM and YoY contribution.
    Thresholds are in percentage points of contribution.
    """
    mom = r["contribution_pp"]
    yoy = r["yoy_contribution_pp"]
    if yoy > yoy_thresh and mom > mom_thresh:
        return "Persistent upward pressure"
    elif yoy > yoy_thresh and mom <= mom_thresh:
        return "Upward pressure but easing"
    elif yoy <= yoy_thresh and mom > mom_thresh:
        return "Short-term upside shock"
    else:
        return "Disinflation / minor"

combined_mom_yoy["regime"] = combined_mom_yoy.apply(classify_row, axis=1)

print("Canada – MoM vs YoY contribution regimes:")
display(
    combined_mom_yoy[[
        "component",
        "contribution_pp",
        "share_of_total_%",
        "yoy_contribution_pp",
        "yoy_share_of_total_%",
        "regime",
    ]].sort_values("yoy_contribution_pp", ascending=False)
)

# -------------------------------------------------------------------
# CANADA – scatter: YoY vs MoM contributions by component
# -------------------------------------------------------------------
fig, ax = plt.subplots(figsize=(8, 8))

x_vals = combined_mom_yoy["yoy_contribution_pp"]
y_vals = combined_mom_yoy["contribution_pp"]

ax.scatter(x_vals, y_vals)

for _, row in combined_mom_yoy.iterrows():
    ax.text(
        row["yoy_contribution_pp"],
        row["contribution_pp"],
        row["component"],
        fontsize=9,
        ha="center",
        va="bottom"
    )

ax.axhline(0, color="black", linewidth=0.8)
ax.axvline(0, color="black", linewidth=0.8)

ax.set_xlabel("YoY contribution to CPI (percentage points)", fontsize=11)
ax.set_ylabel("Latest MoM contribution to CPI (percentage points)", fontsize=11)
ax.set_title(
    f"Canada – MoM vs YoY CPI contribution by component\n"
    f"({FROM_MONTH_CAN} → {TO_MONTH_CAN} vs {YOY_FROM} → {YOY_TO})",
    fontsize=14,
    pad=15,
)

plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()

# ===================================================================
# US SECTION – MoM CONTRIBUTIONS
# ===================================================================

# -------------------------------------------------------------------
# 7. US – load CPI-U Table 1 (Sep 2025 news release)
# -------------------------------------------------------------------
us_raw = pd.read_csv(US_TABLE1_PATH)

# Data starts at row 5; first 10 columns are the ones we need
us = us_raw.iloc[5:, :10].copy()
us.columns = [
    "indent",
    "category",
    "rel_importance",
    "index_sep_2024",
    "index_aug_2025",
    "index_sep_2025",
    "yoy_pct",
    "mom_aug_sep_unadj",
    "mom_sa_prev",
    "mom_sa_aug_sep",
]

# Drop empty rows
us = us[us["category"].notna()].copy()

# Convert numeric columns properly
num_cols = [
    "rel_importance",
    "index_sep_2024",
    "index_aug_2025",
    "index_sep_2025",
    "yoy_pct",
    "mom_aug_sep_unadj",
    "mom_sa_prev",
    "mom_sa_aug_sep",
]
for col in num_cols:
    us[col] = pd.to_numeric(us[col], errors="coerce")

print("US – cleaned CPI table (first 15 rows):")
display(us.head(15))

# -------------------------------------------------------------------
# 8. US – pick major categories and compute contributions
# -------------------------------------------------------------------
us_major_components = [
    "Food",
    "Energy",
    "Shelter",
    "Medical care",
    "Transportation services",
]

us_major = us[us["category"].isin(us_major_components)].copy()

# Convert relative importance from percent to fraction
us_major["weight_frac"] = us_major["rel_importance"] / 100.0

# Contribution to monthly CPI change (Aug → Sep), in percentage points:
us_major["contribution_pp"] = (
    us_major["weight_frac"] * us_major["mom_aug_sep_unadj"]
)

# Share of the US CPI move accounted for by each of these categories
total_contrib_us = us_major["contribution_pp"].sum()
if abs(total_contrib_us) < 1e-9:
    print("Total US CPI move is ~0; shares would be unstable.")
    us_major["share_of_total_%"] = np.nan
else:
    us_major["share_of_total_%"] = (
        us_major["contribution_pp"] / total_contrib_us * 100.0
    )

print("US – contributions to CPI change (Aug → Sep 2025):")
display(us_major)

# -------------------------------------------------------------------
# 8a. US – headline summary + sanity check
# -------------------------------------------------------------------
us_all_items = us[us["category"] == "All items"].iloc[0]
headline_change_us = us_all_items["mom_aug_sep_unadj"]  # already in %

top3_us = us_major.sort_values("contribution_pp", ascending=False).head(3)
top3_us_share = top3_us["share_of_total_%"].sum()

print(
    f"US headline CPI moved {headline_change_us:.2f}% m/m "
    "from August to September 2025."
)
print("Top 3 US drivers:")
for _, row in top3_us.iterrows():
    print(
        f"  - {row['category']}: "
        f"{row['contribution_pp']:.3f} p.p. "
        f"({row['share_of_total_%']:.1f}% of the move)"
    )
print(f"Top 3 explain {top3_us_share:.1f}% of the total move.\n")

print(
    "Sanity check – sum of component contributions (subset of basket): "
    f"{total_contrib_us:.3f} p.p."
)
print(
    f"Headline CPI move from Table 1: {headline_change_us:.3f} p.p.\n"
)

# -------------------------------------------------------------------
# 9. US – bar chart: horizontal, cleaner layout
# -------------------------------------------------------------------
us_plot = us_major.sort_values("contribution_pp")  # smallest at bottom
y_us = np.arange(len(us_plot))

fig, ax = plt.subplots(figsize=(12, 6))
bars = ax.barh(y_us, us_plot["contribution_pp"])

ax.set_yticks(y_us)
ax.set_yticklabels(us_plot["category"], fontsize=9)
ax.set_xlabel("Contribution to monthly CPI change (percentage points)", fontsize=11)
ax.set_title("US: CPI component contributions, Aug → Sep 2025", fontsize=16, pad=15)
ax.tick_params(axis="x", labelsize=10)
ax.axvline(0, color="black", linewidth=0.8)

for bar, share in zip(bars, us_plot["share_of_total_%"]):
    if pd.isna(share):
        continue
    width = bar.get_width()
    x_pos = width + (0.01 if width >= 0 else -0.01)
    va = "center"
    ha = "left" if width >= 0 else "right"
    ax.text(
        x_pos,
        bar.get_y() + bar.get_height() / 2,
        f"{share:.1f}%",
        va=va,
        ha=ha,
        fontsize=9,
    )

plt.tight_layout()
plt.show()

# -------------------------------------------------------------------
# 10. US – pie chart: positive contributors only
# -------------------------------------------------------------------
positive_us = us_major[us_major["contribution_pp"] > 0].copy()
positive_us = positive_us.sort_values("share_of_total_%", ascending=False)

fig, ax = plt.subplots(figsize=(9, 9))
wedges, texts, autotexts = ax.pie(
    positive_us["share_of_total_%"],
    labels=positive_us["category"],
    autopct="%1.1f%%",
    startangle=90,
    counterclock=False,
    textprops={"fontsize": 10},
)
for txt in texts:
    txt.set_fontsize(9)
for at in autotexts:
    at.set_fontsize(10)

ax.set_title(
    "US: share of CPI increase by component\n"
    "(positive contributors, Aug → Sep 2025)",
    fontsize=14,
)
plt.show()

# ===================================================================
# CANADA vs US COMPARISON – FOOD & SHELTER
# ===================================================================

can_latest = can_combined.set_index("component")
us_latest  = us_major.set_index("category")

comp_list = ["Food", "Shelter"]

comparison = pd.DataFrame(
    {
        "Canada (p.p.)": [can_latest.loc[c, "contribution_pp"] for c in comp_list],
        "US (p.p.)": [us_latest.loc[c, "contribution_pp"] for c in comp_list],
        "Canada (% of move)": [can_latest.loc[c, "share_of_total_%"] for c in comp_list],
        "US (% of move)": [us_latest.loc[c, "share_of_total_%"] for c in comp_list],
    },
    index=comp_list,
)

print("Canada vs US – Food & Shelter contributions:")
display(comparison)

ax = comparison[["Canada (% of move)", "US (% of move)"]].plot(
    kind="bar", figsize=(10, 6)
)
plt.ylabel("Share of total CPI move (%)", fontsize=11)
plt.title(
    "Canada vs US – share of CPI move from Food & Shelter\n(Aug → Sep 2025)",
    fontsize=14,
)
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=10)

# Add value labels on top of bars
for p in ax.patches:
    height = p.get_height()
    if np.isnan(height):
        continue
    ax.text(
        p.get_x() + p.get_width() / 2,
        height + 1,
        f"{height:.1f}%",
        ha="center",
        va="bottom",
        fontsize=9,
    )

plt.tight_layout()
plt.show()
```
